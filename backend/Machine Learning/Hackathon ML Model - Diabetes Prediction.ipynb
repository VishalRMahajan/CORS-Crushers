## Importing Libraries

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

## Importing Dataset
[Dataset Link - Kaggle Diabetes Dataset](https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database)

```python
diabetes_dataframe = pd.read_csv('datasets/diabetes.csv')
```

```python
Data = diabetes_dataframe.drop(columns=['Outcome','SkinThickness'], axis =1)
Outcome = diabetes_dataframe['Outcome']
```

```python
Data
```

```python
Outcome
```

### Data Standardization

```python
scaler = StandardScaler()
scaler.fit(Data)

Data= scaler.transform(Data)
```

```python
Data
```

### Data Splitting

```python
Data_train, Data_test, Outcome_train , Outcome_test = train_test_split(Data, Outcome, test_size = 0.1, stratify= Outcome, random_state = 2)
```

### Training the SVM Model

```python
svm_classifier = svm.SVC(kernel='linear')
svm_classifier.fit(Data_train, Outcome_train)
```

#### Model Evaluation

```python
Data_train_prediction = svm_classifier.predict(Data_train)
training_data_accuracy = accuracy_score(Data_train_prediction, Outcome_train)

Data_test_prediction = svm_classifier.predict(Data_test)
test_data_accuracy = accuracy_score(Data_test_prediction, Outcome_test)

print("Training Data Accuracy is :",training_data_accuracy)
print("Testing Data Accuracy is :",test_data_accuracy)
```

#### Testing the SVM Model

```python
input_data = np.array([85,66,29,0,26.6,0.351,31]).reshape(1, -1)
scaler.fit(input_data) 

std_data = scaler.transform(input_data)

prediction = svm_classifier.predict(std_data)

if (prediction[0] == 0):
    print('The person is safe')
else:
    print('The person is may get diabetic')
```

### Training Logistic Model

```python
model = LogisticRegression()
model.fit(Data_train, Outcome_train)
```

#### Model Evaluation

```python
accuracy = model.score(Data_test, Outcome_test)
print("Testing Data Accuracy is :",accuracy)
```

### Testing the Logistic Model

```python
input_data = np.array([0, 150, 72, 0, 33.6, 0.627, 18]).reshape(1, -1)

std_data = scaler.transform(input_data)

prediction = model.predict(std_data)
probability = model.predict_proba(std_data)[:, 1]

if probability < 0.2:
    print("Probability of Having Diabetes is Very Low")
elif probability < 0.4:
    print("LowProbability of Having Diabetes is Low")
elif probability < 0.6:
    print( "Probability of Having Diabetes is Moderate")
elif probability < 0.8:
    print("Probability of Having Diabetes is High")
else:
    print("Probability of Having Diabetes is Very High")
```
